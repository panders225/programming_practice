```sas
/* Question 1 - use proc sql to create a table with selected columns consisting only of 
schools that have 5 or more players in the dataset*/
proc sql;
create table basketball as
select 
baser.seed
, baser.school
, baser.region
, baser.player
, baser.ppg
, baser.rpg
from 
bball.ncaam06 baser

inner join
(
select 
school
, count(distinct player) as tot_players
from
bball.ncaam06
group by 
school
having calculated tot_players >= 5
) sq
on sq.school = baser.school
;
quit;

```

```sas
/* Question 2 - Overhaul the data-driven macro that will produce the region-level reports as the previous assignment, 
but only using sql statements */

%macro looper();

/*2a - create a table with an unduplicated list of the regions*/
proc sql;
create table region_dedup as
select
distinct
region
from
basketball
;

/*2b - assign a macro containing the number of regions from the sqlobs macro value */
%let region_num = &sqlobs.;

/*2c - create macro variables for each region */
data _null_;
set region_dedup;
call symputx(compress('region_num_'||_n_), region);
run;

/* 2d - reproduce the assignment exactly as it appears in the instructions
   2e - use a loop to iterate through these */


    %do i = 1 %to &region_num.;
        title "Team Statistics for the &&region_num_&i Region";
/*         proc sql; */
        select 
        seed 'Row'
        , school 'Team'
        , avg(ppg) as avg_points 'Average Points' format 8.1
        , avg(rpg) as avg_rebounds 'Average Rebounds' format 8.1
        from
        basketball
        where region="&&region_num_&i"
        group by 
        seed
        , school
        ;
        quit;
    %end;
%mend;
%looper();


```

```sas
/* 3. Use an SQL procedure to create a report of the top 20 players with the highest number of points */
title 'Top 20 Scorers';
proc sql;
select 
player 'Name'
, ppg 'Points'
, school 'Team'
, region 'Region'
, seed 'Seed'
from
    (
    select
    *
    from
    basketball
    order by 
    ppg descending
    )
where monotonic() <= 20
;
quit;
```
